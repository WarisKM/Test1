```r
ggplot(sample_n(diamonds,1000),aes(cut, fill = cut))+
geom_bar()+
labs(title = 'comparison in respect to the quality of the cut',
caption = "Source : Diamonds buildin by dplyr library")+
scale_fill_manual(values = c("#f1b3e2" ,"#f4db9a", "#a8fe99" ,"#b3f1e1","#b3c3f1" )) +
theme_minimal()
```

a little bit of box plot for the my love for IQR

---



```r
set.seed(420)
ggplot(sample_n(diamonds,1000),aes(cut, price, col = cut))+
geom_boxplot(outlier.shape = NA)+
labs(title ="Relationship between Cut and Price",
caption ="Source : Diamonds buildin by dplyr library",
x = "cut",
y = "price",
)+
stat_summary(color = "#f1b3e2")+
stat_summary(
fun.min = function(x) mean(x) - sd(x)
,fun.max = function(x) mean(x) + sd(x)
,geom = "errorbar"
,color = "#fc8d59"
,width = 1
) +
theme_minimal()
```

```r
set.seed(420)
ggplot(sample_n(diamonds,1000),
aes(carat,price,color=color))+
geom_point(alpha = 0.6)+
labs(title ="Relationship between carat and price",
caption ="Source : Diamonds buildin by dplyr library")+

theme_minimal()
```

```r
set.seed(420)
ggplot(sample_n(diamonds,1000),aes(price, fill=cut))+
geom_density(alpha=0.1)+
labs(title ="Density between price and cut",
caption ="Source : Diamonds buildin by dplyr library")+
theme_minimal()
```

```r
set.seed(420)
ggplot(sample_n(diamonds,1000),aes(price, carat, group=color, col=color)) +
geom_point(size=3) +
geom_line(size=1.5) +
labs(title ="Relationship between Carat and Price of color",
caption ="Source : Diamonds buildin by dplyr library")+
theme_minimal()
```
